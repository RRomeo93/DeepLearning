## Comparaison Partie 1

```python
import json
import os

configs = ["A", "B", "C1", "C2"]
print(f"{'Config':<10} {'Val Acc':>10} {'Val F1':>10}")
print("-" * 32)
for c in configs:
    path = os.path.join("results", f"part1_{c}_history.json")
    if os.path.exists(path):
        with open(path) as f:
            h = json.load(f)
        print(f"{c:<10} {max(h['val_acc']):>10.4f} {max(h['val_f1']):>10.4f}")
    else:
        print(f"{c:<10} {'—':>10} {'—':>10}")
```
