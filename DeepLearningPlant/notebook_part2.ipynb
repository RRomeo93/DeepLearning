```python
import json
import os

experiments = {
    "Part1 A": "part1_A_history.json",
    "Part1 B": "part1_B_history.json",
    "Part1 C1": "part1_C1_history.json",
    "Part1 C2": "part1_C2_history.json",
    "Part2": "part2_same_res_history.json"
}

print(f"{'Experiment':<12} {'Val Acc':>10} {'Val F1':>10}")
print("-" * 34)
for name, fname in experiments.items():
    path = os.path.join("results", fname)
    if os.path.exists(path):
        with open(path) as f:
            h = json.load(f)
        print(f"{name:<12} {max(h['val_acc']):>10.4f} {max(h['val_f1']):>10.4f}")
    else:
        print(f"{name:<12} {'—':>10} {'—':>10}")
```
