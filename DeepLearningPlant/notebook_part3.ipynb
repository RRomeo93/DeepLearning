## Visualisation des Patch Weight Overlays

Superposition des poids par patch sur l'image originale pour chaque fonction de pondÃ©ration $f(r_p)$.

Fonctions testÃ©es :
- **linear** : $w_p = \varepsilon + r_p$
- **power** : $w_p = (\varepsilon + r_p)^{0.5}$
- **normalized** : $w_p / \text{mean}(w_p)$
- **log** : $w_p = \log(1 + r_p + \varepsilon)$
- **sigmoid** : $w_p = \sigma(\alpha \cdot (r_p - 0.5))$

```python
%matplotlib inline
import os
import torch
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
from PIL import Image
from torchvision import transforms
from patch_weights import compute_patch_ratios, compute_patch_weights
import config


IMG_SIZE = 224
PATCH_SIZE = 16
GRID_SIZE = IMG_SIZE // PATCH_SIZE  


orig_dir = config.ORIG_DIR
seg_dir = config.SEG_DIR


import pandas as pd
df = pd.read_csv(config.CSV_PATH, header=1)
df = df.dropna(subset=["epines"])
df = df[df["epines"].isin([0.0, 1.0])].reset_index(drop=True)

sample_code = None
for idx in range(len(df)):
    code = str(df.iloc[idx]["code"]).replace(".0", "")
    fname = f"{code}.jpg"
    if os.path.isfile(os.path.join(orig_dir, fname)) and os.path.isfile(os.path.join(seg_dir, fname)):
        sample_code = code
        break

print(f"Image utilisee : {sample_code}.jpg")


img_orig = Image.open(os.path.join(orig_dir, f"{sample_code}.jpg")).convert("RGB")
img_seg = Image.open(os.path.join(seg_dir, f"{sample_code}.jpg")).convert("RGB")


resize_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE)),
    transforms.ToTensor(),
])

img_orig_t = resize_tf(img_orig)  
img_seg_t = resize_tf(img_seg)    


mask_pil = img_seg.convert("L").point(lambda p: 255 if p > 10 else 0)
mask_tf = transforms.Compose([
    transforms.Resize((IMG_SIZE, IMG_SIZE), interpolation=transforms.InterpolationMode.NEAREST),
    transforms.ToTensor(),
])
mask = mask_tf(mask_pil).unsqueeze(0)  


img_display = img_orig_t.permute(1, 2, 0).numpy()


weight_functions = {
    "linear":     {"method": "linear"},
    "power":      {"method": "power", "gamma": 0.5},
    "normalized": {"method": "normalized"},
    "log":        {"method": "log"},
    "sigmoid":    {"method": "sigmoid", "gamma": 10.0},
}


all_weights = {}
for name, params in weight_functions.items():
    w = compute_patch_weights(
        mask, PATCH_SIZE,
        method=params["method"],
        gamma=params.get("gamma", None)
    )
    
    w_norm = w / (w.mean() + 1e-8)
    all_weights[name] = w_norm[0].numpy()  

print(f"Poids calcules pour {len(all_weights)} fonctions")
for name, w in all_weights.items():
    print(f"  {name:>12s} : min={w.min():.3f}  max={w.max():.3f}  mean={w.mean():.3f}")
```

```python


n_funcs = len(all_weights)
fig, axes = plt.subplots(1, n_funcs + 1, figsize=(4 * (n_funcs + 1), 4))


axes[0].imshow(img_display)
axes[0].set_title("Non-segmented (resized)", fontsize=10)
axes[0].axis("off")


all_vals = np.concatenate([w for w in all_weights.values()])
vmin_global = 0
vmax_global = np.percentile(all_vals, 98)  


for idx, (name, weights) in enumerate(all_weights.items()):
    ax = axes[idx + 1]
    w_grid = weights.reshape(GRID_SIZE, GRID_SIZE)
    
   
    ax.imshow(img_display, alpha=0.4)
    
 
    cmap = plt.cm.jet
    norm = mcolors.Normalize(vmin=vmin_global, vmax=vmax_global)
    
    for row in range(GRID_SIZE):
        for col in range(GRID_SIZE):
            weight_val = w_grid[row, col]
            color = cmap(norm(weight_val))
            
         
            rect = plt.Rectangle(
                (col * PATCH_SIZE, row * PATCH_SIZE),
                PATCH_SIZE, PATCH_SIZE,
                linewidth=0.5,
                edgecolor='white',
                facecolor=color,
                alpha=0.6
            )
            ax.add_patch(rect)
    
    ax.set_xlim(0, IMG_SIZE)
    ax.set_ylim(IMG_SIZE, 0)
    ax.set_title(f"{name}\n{GRID_SIZE}x{GRID_SIZE} patches", fontsize=10)
    ax.axis("off")


sm = plt.cm.ScalarMappable(cmap=cmap, norm=norm)
sm.set_array([])
cbar = fig.colorbar(sm, ax=axes.tolist(), shrink=0.8, pad=0.02)
cbar.set_label("Patch weight (normalized, mean=1)", fontsize=9)

plt.suptitle(f"Patch weights overlays (img_size={IMG_SIZE}, patch_size={PATCH_SIZE})\n{sample_code}.jpg",
             fontsize=12, y=1.02)
plt.tight_layout()
plt.savefig("results/part3_patch_weight_overlays.png", dpi=150, bbox_inches="tight")
print("Sauvegarde: results/part3_patch_weight_overlays.png")
plt.show()
```
